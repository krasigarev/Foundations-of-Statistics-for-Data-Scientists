# PYTHON FOR SAMPLING DISTRIBUTIONS

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

import statistics
```

## Simulation to Illustrate a Sampling Distribution

```python
n, p = 2271, 0.50

x = np.random.binomial(n, p, 1)
print(x), print(x/n)
```

```python
results = np.random.binomial(n, p, 1000000) / n 
statistics.mean(results)
```

```python
statistics.stdev(results)
```

```python
plt.hist(results, bins=14, edgecolor='k')
plt.xlabel('Sample proportion')
plt.ylabel('Frequency');
```

## Law of Large Numbers

```python
n1, n2, n3 = 10, 1000, 1000000
y1 = np.random.uniform(0, 100, n1)
mean1 = sum(y1)/n1
print(mean1)
```

```python
y2 = np.random.uniform(0, 100, n2)
mean2 = sum(y2)/n2
print(mean2)
```

```python
y3 = np.random.uniform(0, 100, n3)
mean3 = sum(y3)/n3
print(mean3)
```
