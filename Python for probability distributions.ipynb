# PYTHON FOR PROBABILITY DISTRIBUTIONS

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import uniform
```

## Binomial Distribution

```python
binom.pmf(1, 12, 0.20)
```

```python
fig, ax = plt.subplots(1, 1)
n, pi = 12, 0.2
y = list(range(0, 13))
ax.vlines(y, 0, binom.pmf(y, n, pi), color='b', lw=5, alpha=0.5)
plt.xlabel('y')
plt.ylabel('P(y)')
plt.xticks(np.arange(min(y), max(y)+1, 1.0))

display(list(binom.pmf(y, n, pi)))
```

```python
mean, variance, skewness = binom.stats(n, pi, moments='mvs')
print(mean.round(2), variance.round(2), skewness.round(8))
```

```python
fig, ax = plt.subplots(1,1)
n, p = 3, 0.5
x = [0,1,2,3]
ax.vlines(x, 0, binom.pmf(x, n, p), color='b', lw=5, alpha=0.5)
plt.xlabel('y')
plt.ylabel('P(y)')
plt.xticks(np.arange(min(x), max(x)+1, 1));
```

## Poisson Distribution

```python
poisson.pmf(0, 2.3)
```

```python
poisson.cdf(130, 100) - poisson.cdf(69, 100)
```

```python
poisson.cdf(120, 100) - poisson.cdf(79, 100)
```

## Python Functions for Continuous Probability Distributions

### Uniform Distribution

```python
fig, ax = plt.subplots(1,1)

x = np.linspace(uniform.ppf(0.01), uniform.ppf(0.99), 100)
rv = uniform()

ax.plot(x, rv.pdf(x), lw=2, color='b')
plt.plot([-0.3, 0], [0, 0], lw=2, color='r')
plt.plot([1, 1.3], [0, 0], lw=2, color='g')
plt.xticks(np.arange(0, 1.2, 0.2));
```

```python
fig, ax = plt.subplots(1,1)

x = np.linspace(uniform.ppf(0.01), uniform.ppf(0.99), 100)
rv = uniform()

ax.plot(x, rv.cdf(x), lw=2, color='b')
plt.plot([-0.3, 0], [0, 0], lw=2, color='r')
plt.plot([1, 1.3], [1, 1], lw=2, color='g')
plt.xticks(np.arange(0, 1.2, 0.2));
```
